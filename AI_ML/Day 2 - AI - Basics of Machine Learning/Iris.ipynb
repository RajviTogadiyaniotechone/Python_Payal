```python
cols = df.columns
print(cols)
```

```python
cols = cols[1:5]
# columns with the integer values
```

```python
df1 = df[cols]
# will save it from new dataframe variable
```

```python
newcols = {
    "sepal.length" : "sepallength",
    "sepal.width" : "sepalwidth",
    "petal.length" : "petallength",
    "petal.width" : "petalwidth"
}

df.rename(columns = newcols, inplace = True)

print(df.head())
```
