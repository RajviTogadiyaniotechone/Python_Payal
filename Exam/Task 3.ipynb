```python
import cv2
print(cv2.__version__)
```

```python
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('Image.jpg')
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Apply noise reduction (GaussianBlur)
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
img_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Edge detection using Canny
edges = cv2.Canny(blurred_image, 100, 200)
img_rgb = cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Find contours in the edge-detected image
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for contour in contours:
    epsilon = 0.02 * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)

    if len(approx) == 4:  # A rectangle contour is likely the license plate
        x, y, w, h = cv2.boundingRect(approx)
        license_plate_image = gray_image[y:y+h, x:x+w]
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
```
