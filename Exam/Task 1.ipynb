```python
##2.	Clean the data by handling missing values and removing duplicates.
df_clean = df.dropna(subset = ['Quantity Sold','Total Revenue'])
df_clean = df_clean.drop_duplicates()
print(df_clean.head(10))
```

```python
df.info()
```

```python
##3.	Calculate and display:
##o	Total sales per product.
##o	Total revenue per region.
##o	The month with the highest sales.

a = df.groupby('Product')['Quantity Sold'].sum()
print("Total Sales per Product:\n",a)

b = df.groupby('Region')['Total Revenue'].sum()
print("Total revenue per region::",b)

df['month'] = pd.DatetimeIndex(df['Date']).month

c = df.groupby('month')['Quantity Sold'].sum()
print("total",c)

Highest_sales = c.idxmax()
print("highestsales month::::",Highest_sales)

```

```python
#o	Sales trends over time using Matplotlib or Seaborn.

import matplotlib.pyplot as plt
import pandas as pd
import datetime

df['Date'] = pd.DatetimeIndex(df['Date'])

plt.figure(figsize=(10, 6)) 
plt.plot(df['Date'], df['Quantity Sold'], marker='o', color='b', linestyle='-', label='Quantity Sold')

plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)  
plt.grid(True)
plt.tight_layout()

plt.show()                        
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting Sales Trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Quantity Sold', marker='o')
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Top 5 products by revenue
total_revenue_product = df.groupby('Product')['Total Revenue'].sum()
top_5_products = total_revenue_product.nlargest(5)

# Plotting Top 5 Products by Revenue
plt.figure(figsize=(8, 5))
top_5_products.plot(kind='bar', color='skyblue')
plt.title('Top 5 Products by Revenue')
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```
